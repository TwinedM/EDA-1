```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# univariate only
sns.displot(tips['total_bill'],kde = False, bins = 40)
plt.show()
```

```python
# Jointplot - matches up 2 distplots for bivariate data and with your choice of comparitive tool
sns.jointplot(x ='total_bill' , y ='tip',data = tips , kind = 'reg')
plt.show()
```

```python
# pairplot - all types of data frames would be created as in pairwise (good to get an idea of the data) 
sns.pairplot(tips,hue = 'sex', palette = 'coolwarm')
plt.show()
```

```python
 # rugplots - dash mark for every point on a univariate dist (KDE are built over them)
sns.rugplot(tips['total_bill'])
plt.show()
```

```python
sns.barplot(x = 'sex', y = 'total_bill' , data = tips)
plt.show()
```

```python
import numpy as np
# estimator - what function do you want in this is this
sns.barplot(x = 'sex' , y = 'total_bill' , data = tips , estimator = np.std)
plt.show()
```

```python
## Grid - They just make a grid of plots which u can customise and use for data
iris = sns.load_dataset('iris')
iris.head()

```

```python
## Pair grid - gives me an empty canvas for as a grid
sns.PairGrid(iris)
plt.show()
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
plt.show()
```

```python
m = sns.PairGrid(iris)
m.map_diag(plt.hist)
m.map_upper(plt.scatter)
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(data = titanic , x = 'fare' , y = 'age' , kind = 'scatter')
plt.show()
```

```python
sns.displot(data = titanic , x = 'fare')
plt.show()
```

```python
sns.boxplot(x = 'class' , y = 'age' , data = titanic , width = 0.4 , palette = 'coolwarm'  )
plt.show()
```

```python
sns.swarmplot(x = 'class' , y = 'age' , data = titanic , size = 4 , palette = 'muted' , hue = 'class')
plt.show()
```

```python
sns.countplot(x='sex',data=titanic)
plt.show()
```

```python
g = sns.FacetGrid(data=titanic,col='sex')
g.map(plt.hist,'age')
plt.show()
```
